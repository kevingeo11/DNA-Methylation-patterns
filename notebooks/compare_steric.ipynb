### Steric Compare

```python
import pickle
```

```python
nuc_structure = ['1KX5', '7V9J']
dmnts = ['3PTA', '7XIB']
```

```python
x_y_clash_dict = {}
for _nuc in nuc_structure:
    for _dnmt in dmnts:
        infile = main_path / 'superposition' / f'clash_{_nuc}_{_dnmt}' / 'x_y_dict'
        with open(infile, 'rb') as fin:
            if _nuc not in x_y_clash_dict:
                x_y_clash_dict[_nuc] = {}
            x_y_clash_dict[_nuc][_dnmt] = pickle.load(fin)
```

```python
fig = plt.figure(figsize=(18, 7))
ax = plt.subplot(111)

_nuc = nuc_structure[0]

for _dnmt in x_y_clash_dict[_nuc]:
    x = list(x_y_clash_dict[_nuc][_dnmt].keys())
    y = list(x_y_clash_dict[_nuc][_dnmt].values())

    ax.plot(x, y, linestyle="-", marker="o", linewidth=2, label=_dnmt)

    # if _dnmt == '7XIB':
    #     for pos in range(0,len(x)):
    #         ax.text(x[pos]+0.2, y[pos]+0.7, x[pos], fontsize=12)
    
plt.xlim(0,140)
plt.xticks(range(0,len(x)+5,5), fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Nucleosome position [bp]", fontsize=18)
plt.ylabel("DNMT1 atoms clashing [%]", fontsize=18)
plt.ylim(top=100)

ax.axvline(x=74,c="grey",linewidth=2,linestyle="--")#4169e1

ax.legend(fontsize=16)
```
